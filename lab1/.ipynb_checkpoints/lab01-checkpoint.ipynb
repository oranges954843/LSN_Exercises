## **Exercise 1** 

- Let's test the Pseudo-Random Number generator downloaded from ariel. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("ave.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,10000],[0.5,0.5],label = 'y=1/2')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()

plt.show()
```

2. Results of the estimation of $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$ and its statistical uncertainty with data blocking.

```python
x, f, error = np.loadtxt("sigma2.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,10000],[1/12,1/12],label = 'y=1/12')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()

plt.show()
```

3. Results for the $\chi^2$ test.

```python
x, f = np.loadtxt("chi2.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f)
plt.plot([0,100],[100,100],label = 'y=100')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()

plt.show()
```
