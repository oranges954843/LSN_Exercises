# The Travelling Salesman Problem with a Simulated Annealing algorithm

- In the following are shown pictures of the best cost function and of the best path. The SA algorithm has been implemented embedding the system in a ficticious temperature from $T = 0.8$ to $0.1$ (decreasing it with $T_{new} = T_{old} - 0.05$). At every temperature have been performed $M = 8000$ Monte Carlo steps divided in $N = 10$ blocks in order to use blocking average at every temperature.

### The Travelling Salesman on a circle

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f = np.loadtxt("best_CostFunction_circle.out", usecols=(0,1), delimiter='   ', unpack='true')

D = 120000 #number of total MC step
M = 8000 #number of MC step at every temperature
L = 800 #number of MC step in every block
N = int(M/L) # number of blocks at every temperature
K = int(D/M)
ave = np.zeros(N)
ave2 = np.zeros(N)

sigma = np.zeros(K)
ave_final = np.zeros(K)

for k in range(K):
    ave_prog = 0
    ave2_prog = 0
    for i in range(N):
        sum = 0
        for j in range(L):
            sum += f[j+i*L+k*M]
        ave[i] = sum/L
        ave2[i] = ave[i]*ave[i]
        ave_prog += ave[i]
        ave2_prog += ave2[i]

    ave_prog /= N
    ave2_prog /= N
    sigma[k] = math.sqrt((ave2_prog-pow(ave_prog,2))/(N))
    ave_final[k] = ave_prog



T = np.zeros(K) 
for i in range(K):
    T[i] = 0.8 - 0.05 * i
    

plt.errorbar(T,ave_final,yerr=sigma)
plt.xlabel('Temperature')
plt.ylabel('L$^2$')
plt.grid('true')
plt.title('Best $L^2$ on the circle vs Temperature')

plt.show()

p, x = np.loadtxt("best_path_circle.out", usecols=(0,1), delimiter='  ', unpack='true')

plt.plot(x,p)
plt.xlabel('x')
plt.ylabel('Cities')
plt.grid('true')
plt.title('Best path on the circle')

plt.show()
```

### The Travelling Salesman in a square

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f = np.loadtxt("best_CostFunction_square.out", usecols=(0,1), delimiter='   ', unpack='true')


D = 120000 #number of total MC step
M = 8000 #number of MC step at every temperature
L = 800 #number of MC step in every block
N = int(M/L) # number of blocks at every temperature
K = int(D/M)
ave = np.zeros(N)
ave2 = np.zeros(N)

sigma = np.zeros(K)
ave_final = np.zeros(K)

for k in range(K):
    ave_prog = 0
    ave2_prog = 0
    for i in range(N):
        sum = 0
        for j in range(L):
            sum += f[j+i*L+k*M]
        ave[i] = sum/L
        ave2[i] = ave[i]*ave[i]
        ave_prog += ave[i]
        ave2_prog += ave2[i]

    ave_prog /= N
    ave2_prog /= N
    sigma[k] = math.sqrt((ave2_prog-pow(ave_prog,2))/(N))
    ave_final[k] = ave_prog



T = np.zeros(K)
for i in range(K):
    T[i] = 0.8 - 0.05 * i
    

plt.errorbar(T,ave_final,yerr=sigma)
plt.xlabel('Temperature')
plt.ylabel('L$^2$')
plt.grid('true')
plt.title('Best $L^2$ in the square vs Temperature')

plt.show()

p, x = np.loadtxt("best_path_square.out", usecols=(0,1), delimiter='  ', unpack='true')

plt.plot(x,p)
plt.xlabel('x')
plt.ylabel('Cities')
plt.grid('true')
plt.title('Best path in the square')

plt.show()
```

#### Observations

- The TSP on the circle: in the GA algorithm the best $L^2$ in the population oscillates (after some generations) more or less in the range $(1.0,2.0)$ whereas for the SA algorithm $L^2$ decreases reaching more or less $3.0$. This can be due to the presence of a minimum in the SA simulation higher than the one in the GA simulation.

- The TSP in the square: in the GA algorithm the best $L^2$ in the population oscillates more or less in the range $(5.0,7.0)$ and for the SA algorithm $L^2$ decreases below $6.0$. This shows that both algorithms have reached more or less the same minimum.


# The TSP in a square with four parallel GA searches 

- Comparison of the $L^2$ averaged on the best half of the population between GA without MPI and GA with MPI vs number of generations.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f = np.loadtxt("ave_CostFunction_MPI_1.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA with MPI')
x, f = np.loadtxt("ave_CostFunction_MPI_2.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA with MPI')

x, f = np.loadtxt("ave_FXY_GA.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA without MPI')

plt.xlabel('# of generations')
plt.ylabel('<L$^2$>')
plt.grid('true')
plt.title('$L^2$ averaged on the best half vs generations')
plt.legend()

plt.show()

x, f = np.loadtxt("ave_CostFunction_MPI_3.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA with MPI')
x, f = np.loadtxt("ave_CostFunction_MPI_4.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA with MPI')

x, f = np.loadtxt("ave_FXY_GA.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA without MPI')

plt.xlabel('# of generations')
plt.ylabel('<L$^2$>')
plt.grid('true')
plt.title('$L^2$ averaged on the best half vs generations')
plt.legend()

plt.show() 
```

- Comparison of the best $L^2$ of the population between GA without MPI and GA with MPI vs number of generations.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f = np.loadtxt("best_CostFunction_MPI_1.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA with MPI')
x, f = np.loadtxt("best_CostFunction_MPI_2.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA with MPI')

x, f = np.loadtxt("best_FXY_GA.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA without MPI')

plt.xlabel('# of generations')
plt.ylabel('Best $L^2$')
plt.grid('true')
plt.title('Best $L^2$ vs generations')
plt.legend()

plt.show()

x, f = np.loadtxt("best_CostFunction_MPI_3.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA with MPI')
x, f = np.loadtxt("best_CostFunction_MPI_4.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA with MPI')

x, f = np.loadtxt("best_FXY_GA.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f,label = 'GA without MPI')

plt.xlabel('# of generations')
plt.ylabel('Best $L^2$')
plt.grid('true')
plt.title('Best $L^2$ vs generations')
plt.legend()

plt.show() 
```

- We can see how the GA algorithm with 4 parallel indipendent optimizations (in contrast with the GA without parallelization) tends to reach lower values of the best $L^2$ and of the average $<L^2>$ on the best half.
