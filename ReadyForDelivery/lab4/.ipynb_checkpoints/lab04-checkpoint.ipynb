# Verlet Algorithm applied to simulate Argon

## Equilibration

#### Solid phase

Total energy, potential energy, kinetic energy (per particle) and temperature are shown below. The solid phase has been equilibrated with 5 simulations with 1000 steps each. They are shown in Lennard-Jones units. Measurements have been performed every 10 moves.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.arange(500)
x = x * 10
f = np.loadtxt("solid_etot.istant", delimiter = '\n', unpack='true')
plt.plot(x,f, label = '$E/N$')
f = np.loadtxt("solid_epot.istant", delimiter = '\n', unpack='true')
plt.plot(x,f, label = '$U/N$')
f = np.loadtxt("solid_ekin.istant", delimiter = '\n', unpack='true')
plt.plot(x,f, label = '$K/N$')
plt.xlabel('# of time steps')
plt.ylabel('Energy')
plt.grid('true')
plt.legend()
plt.title('$E/N$, $U/N$, $K/N$ equilibration')

plt.show()
```

```python
f = np.loadtxt("solid_temp.istant", delimiter = '\n', unpack='true')
plt.plot(x,f)
plt.xlabel('# of time steps')
plt.ylabel('Temperature')
plt.grid('true')
plt.title('Solid phase temperature equilibration')

plt.show()
```

## MD simulation of Argon in three different phases

- Solid phase

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

e_f = 388.68 # m*sigma**2 / dt**2

x = np.arange(25)
x *= 10000
f, error = np.loadtxt("solid_Ar_etot.out", usecols=(1,2), delimiter='   ', unpack='true')
f = f * e_f
error = error * e_f
plt.errorbar(x,f,yerr=error)
plt.xlabel('# of time steps')
plt.ylabel('$E/N$ (amu*$nm^2$/$fs^2$)')
plt.grid('true')
plt.title('Total internal energy per particle')

plt.show()

f, error = np.loadtxt("solid_Ar_epot.out", usecols=(1,2), delimiter='   ', unpack='true')
f = f * e_f
error = error * e_f
plt.errorbar(x,f,yerr=error)
plt.xlabel('# of time steps')
plt.ylabel('$U/N$ (amu*$nm^2$/$fs^2$)')
plt.grid('true')
plt.title('Potential energy per particle')

plt.show()

f, error = np.loadtxt("solid_Ar_ekin.out", usecols=(1,2), delimiter='   ', unpack='true')
f = f * e_f
error = error * e_f
plt.errorbar(x,f,yerr=error)
plt.xlabel('# of time steps')
plt.ylabel('$K/N$ (amu*$nm^2$/$fs^2$)')
plt.grid('true')
plt.title('Kinetic energy per particle')

plt.show()

f, error = np.loadtxt("solid_Ar_temp.out", usecols=(1,2), delimiter='   ', unpack='true')
f = f * 120
error = error * 120
plt.errorbar(x,f,yerr=error)
plt.xlabel('# of time steps')
plt.ylabel('T (K)')
plt.grid('true')
plt.title('Solid phase temperature')

plt.show()
```
