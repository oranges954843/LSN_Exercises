# Exercise 1

The following parameters have been used in order to compute at time $t=0$ the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$

- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$


1. By sampling directly the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$ with 16 thousand MC steps divided in 80 blocks.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("call_final.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,16000],[14.9758,14.9758],label = 'y=14.9758')
plt.xlabel('# of MC steps')
plt.ylabel('call option price')
plt.grid('true')
plt.legend()
plt.title('European call option price sampling the final asset')

plt.show()

x, f, error = np.loadtxt("put_final.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,16000],[5.4595,5.4595],label = 'y=5.4595')
plt.xlabel('# of MC steps')
plt.ylabel('put option price')
plt.grid('true')
plt.legend()
plt.title('European put option price sampling the final asset')

plt.show()
```

2. By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals and dividing the 13 thousand MC steps in 100 blocks.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("call_GBM.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,13000],[14.9758,14.9758],label = 'y=14.9758')
plt.xlabel('# of MC steps')
plt.ylabel('call option price')
plt.grid('true')
plt.legend()
plt.title('European call option price sampling the path')

plt.show()

x, f, error = np.loadtxt("put_GBM.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,13000],[5.4595,5.4595],label = 'y=5.4595')
plt.xlabel('# of MC steps')
plt.ylabel('put option price')
plt.grid('true')
plt.legend()
plt.title('European put option price sampling the path')

plt.show()
```
