# Sampling Hydrogen atom probability density with Metropolis

1. A unifrom transition probability is used in the Metropolis algorithm to sample $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$ in Cartesian coordinates.

- Estimation of $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and its statistical uncertainty by means of $10^6$ MC steps divided in 100 blocks. The estimation has been performed after having equilibrated (starting from (0,0,0)) with 2000 steps.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("output_GS.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,100],[1.5,1.5],label = 'y=1.5')
plt.xlabel('# of blocks')
plt.ylabel('<R>/$a_0$')
plt.grid('true')
plt.legend()
plt.title('Estimation of r with $\Psi_{1,0,0}(x,y,z)$')

plt.show()
```

- Estimation of $\left\langle r \right\rangle_{\Psi_{2,1,0}}$ and its statistical uncertainty by means of $10^6$ MC steps divided in 100 blocks. The estimation has been performed after having equilibrated (starting from (1,1,1)) with 2000 steps.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("output_2P.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,100],[5,5],label = 'y=5')
plt.xlabel('# of blocks')
plt.ylabel('<R>/$a_0$')
plt.grid('true')
plt.legend()
plt.title('Estimation of r with $\Psi_{2,1,0}(x,y,z)$')

plt.show()
```

- Some observations.

    1. The more the starting point is far from the origin the more the algorithm will take to converge to an equilibrium configuration. However if the starting point is too far away from the origin the probability of acceptance becomes negligible and so the algorithm will never converge.
    2. The origin is a reasonable starting point for the ground state but it is not for the excited state. Indeed, in $\Psi_{2,1,0}(x,y,z)$, on one hand the $\cos(\theta)$ is not defined at the origin and on the other hand when it is computed the ratio between the probability in the new and old configurations we are dividing by zero.
    3. The MC steps in each block should be large enough to avoid autocorrelation between to successive steps.
    4. If a multivariate normal transition probability is used in place of the uniform transition probability the results are equivalent.
