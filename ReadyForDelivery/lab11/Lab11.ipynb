# First polinomial fit with a Neural Network (NN)

Let's try to fit 

$$
f(x)=4-3x-2x^2+3x^3
$$

for $x \in [-1,1]$.

Parameters for an ideal third order polinomial

```python
# target parameters of f(x) = ax*** + bx** + cx + d
a = 3 
b = -2 
c = -3
d = 4
```

Let's generate a set of input data which will slightly deviate from our ideal behaviour using a random noise set to zero

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 800)
x_valid = np.random.uniform(-1, 1, 80)
x_valid.sort()
y_target = a * pow(x_valid,3) + b * pow(x_valid,2) + c * x_valid + d # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(a * pow(x_train,3) + b * pow(x_train,2) + c * x_train + d, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(a * pow(x_valid,3) + b * pow(x_valid,2) + c * x_valid + d, sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

Let's start using a simple sequential model with just one layer on one neuron only!

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(10, input_shape=(1,), activation='relu'))
model.add(Dense(14, activation='relu'))
model.add(Dense(18, activation='relu'))
model.add(Dense(22, activation='relu'))

model.add(Dense(1))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model.summary()
```

Let's train our model, that is we feed the neuron with the set of training pair x, y_train from which the optimizer will find the best weights to minimize the Mean Square Error loss function (out linear regression function).

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=40,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()

# summarize history for accuracy
#plt.plot(history.history['categorical_accuracy'])      
plt.plot(history.history['mse']) #for generic 'accuracy'
plt.plot(history.history['val_mse'])  #plt.plot(history.history['val_acc'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-2, 2, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

- We can observe that inside the range [-1,1] the NN predict quite well the model but outside (expecially for $x > 1$) the model needs to be improved.

## Conclusions

- Increasing the number of train and valid points improves the fit but after a certain threshold one does not note improvements.

- A noise different from zero but in a narrow range around 0.2 seems to give good results but as one increase the noise, make a bad prediction.

- If one choose a NN with six or more dense layers with more than 10 neurons each, start to over fit the model, which means that one is no more able to add new informations in order to improve the prediction.

- Different activation functions from relu seems to give worse results.

- Changing the optimizer from SGD to Adagrad or Adadelta worsen the result. For example we have:

# Fit of a 2D trigonometric function

Let's extend the model to fit 

$$f(x,y) = \sin(x^2+y^2)$$ 

in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

```python
import numpy as np
import math

n_train = 2500
n_valid = 250

x_train = np.zeros((n_train, 2))
x_valid = np.zeros((n_valid, 2))

np.random.seed(0)
x_train[:,0] = np.random.uniform(-1.5, 1.5, n_train)
x_train[:,1] = np.random.uniform(-1.5, 1.5, n_train)
x_valid[:,0] = np.random.uniform(-1.5, 1.5, n_valid)
x_valid[:,1] = np.random.uniform(-1.5, 1.5, n_valid)

x_valid[:,0].sort()
x_valid[:,1].sort()

f_target = np.sin(x_valid[:,0]**2 + x_valid[:,1]**2)
f_train = np.zeros(n_train)
f_valid = np.zeros(n_valid)

sigma = 0.2

f_train = np.random.normal(np.sin(pow(x_train[:,0],2)+pow(x_train[:,1],2)), sigma)
f_valid = np.random.normal(np.sin(pow(x_valid[:,0],2)+pow(x_valid[:,1],2)), sigma)
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

X=x_valid[:,0]
Y=x_valid[:,1]
Z=f_valid
ax.scatter(X, Y, Z, c='r', marker='o')

X=x_valid[:,0]
Y=x_valid[:,1]
X, Y = np.meshgrid(X, Y)
Z = np.sin(X**2 + Y**2)
ax.plot_surface(X, Y, Z)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(10, 30)
plt.show()
```

```python
model = Sequential()

model.add(Dense(8, input_shape=(2,), activation='relu'))

model.add(Dense(16, input_shape=(1,), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(48, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(80, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
model.summary()
```

```python
history = model.fit(x=x_train, y=f_train, 
          batch_size=32, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid,f_valid))
```

```python
for l in model.layers:
    w = l.get_weights()
    print("layer nodes weights: ", w[0].shape)
    print("layer bias weights: ", w[1].shape)
```

```python
score = model.evaluate(x_valid, f_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
