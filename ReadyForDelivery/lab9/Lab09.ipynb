# Genetic Algorithm to solve the Travelling Salesman Problem 

### Travelling Salesman Problem on a circle

- The TSP is implemented choosing randomly 32 cities placed in [0,1] with periodic boundary conditions and creating an initial population of 1000 individuals and finally optimizing for 250 generations with a genetic algorithm.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f = np.loadtxt("best_CostFunction_circle.out", usecols=(0,1), delimiter='   ', unpack='true')

plt.plot(x,f)
plt.xlabel('# of generations')
plt.ylabel('L$^{(2)}$')
plt.grid('true')
plt.title('Best $L^{(2)}$ vs generations')

plt.show()

x, f = np.loadtxt("ave_CostFunction_circle.out", usecols=(0,1), delimiter='   ', unpack='true')

plt.plot(x,f)
plt.xlabel('# of generations')
plt.ylabel('L$^{(2)}$')
plt.grid('true')
plt.title('$L^{(2)}$ of the best half of the population vs generations')

plt.show()
```

- Note: The best path on the circle is obtained projecting every point of the figure on the abscissa. For practical matters the points in the figure are spread along the ordinate in relation with the city labels

```python
p, x = np.loadtxt("best_path_circle.out", usecols=(0,1), delimiter='  ', unpack='true')

plt.plot(x,p)
plt.xlabel('x')
plt.ylabel('city labels')
plt.grid('true')
plt.title('Best path on the circle')

plt.show()
```

### Travelling Salesman Problem in a square

- The TSP is implemented choosing randomly 32 cities in a box [0,1] x [0,1] and creating an initial population of 1000 individuals and then optimizing for 250 generations with a genetic algorithm.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f = np.loadtxt("best_CostFunction_square.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f)

plt.xlabel('# of generations')
plt.ylabel('L$^{(2)}$')
plt.grid('true')
plt.title('Best $L^{(2)}$ vs generations')


plt.show()

x, f = np.loadtxt("ave_CostFunction_square.out", usecols=(0,1), delimiter='   ', unpack='true')

plt.plot(x,f)
plt.xlabel('# of generations')
plt.ylabel('L$^{(2)}$')
plt.grid('true')
plt.title('$L^{(2)}$ of the best half vs generations')

plt.show()
```

- Picture of the best path in the square

```python
p, x, y = np.loadtxt("best_path_square.out", usecols=(0,1,2), delimiter='  ', unpack='true')

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid('true')
plt.title('Best path in the square')

plt.show()
```
