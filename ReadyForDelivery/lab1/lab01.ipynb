## **Exercise 1** 

1. Results of the estimation of $\langle r \rangle = \int_0^1 r dr = 1/2$ and its statistical uncertainty with data blocking, with 200 Monte Carlo steps per block for 50 blocks. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("ave.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,10000],[0.5,0.5],label = 'y=1/2')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()
plt.title('Estimation of <r> with L = 200, N = 50')

plt.show()
```

2. Results of the estimation of $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$ and its statistical uncertainty with data blocking, with 200 Monte Carlo steps per block for 50 blocks. 

```python
x, f, error = np.loadtxt("sigma2.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,10000],[1/12,1/12],label = 'y=1/12')
plt.xlabel('# of throws')
plt.ylabel('$\sigma^2$')
plt.grid('true')
plt.legend()
plt.title('Estimation of $\sigma^2$ with L = 200, N = 50')

plt.show()
```

3. Results for the $\chi^2$ test.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f = np.loadtxt("chi2.out", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x,f)
plt.plot([0,100],[100,100],label = 'y=100')
plt.xlabel('# of tests')
plt.ylabel('$\chi^2$')
plt.grid('true')
plt.legend()
plt.title('100 $\chi^2$ tests with $10^4$ throws each')

plt.show()
```
