## 1D integral via Monte Carlo

- Let's compute the following 1D integral via Monte Carlo 

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. Results by sampling the integral with a uniform distribution in the interval [0,1] with data blocking for the statistical uncertainties.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("Uniform.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,12000],[1.0,1.0],label = 'y=1')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()

plt.show()
```

2. Results by sampling the integral with a non-uniform distribution in the interval [0,1] with data blocking for the statistical uncertainties. In order to change the probability density to sample [0,1] it has been first made a change in the integrand with an integration by parts.

$$\int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = \int_0^1 \frac{\pi^2}{4}\sin(\pi x/2) x dx$$

After that it is calculated and normalized the probability density (computed considering the Taylor expansion of the $\sin(x)$ function in the integrand arrested at the first order)

$$p(x) = 3 x ^ 2$$

thus, the integral becomes

$$I = \int_0^1 p(x)\frac{\pi ^2}{4}\sin(\pi x/2)\frac{x}{3x^2} dx = 1$$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("NonUniform.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,10000],[1.0,1.0],label = 'y=1')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()

plt.show()
```

## 3D Random Walks in Z3 and R3 

- Random Walk in a cubic lattice with lattice constant $a=1$. It is shown a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle}$, with its statistical uncertainty (computed dividing $10^4$ RWs in 20 blocks), as function of the step $i\in [0,10^2]$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("Z3_lattice.out", usecols=(0,1,2), delimiter='   ', unpack='true')

plt.errorbar(x,f,yerr=error)
plt.xlabel('# of time steps')
plt.ylabel('$\sigma$')
plt.grid('true')
plt.title('Fluctuations respect to the inital position')

plt.show()


plt.plot(x, error)
plt.xlabel('# of time steps')
plt.ylabel('error')
plt.grid('true')
plt.title('Statistical uncertainty of the fluctuations')
plt.show()
```

- Random Walk in the continuum where at each discrete time the walker makes a step of length equal to $a = 1 $ along a random direction. It is shown a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$, with its statistical uncertainty (computed dividing $10^4$ RWs in 20 blocks), as a function of the step $i\in [0,10^2]$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("R3_lattice.out", usecols=(0,1,2), delimiter='   ', unpack='true')

plt.errorbar(x,f,yerr=error)
plt.xlabel('# of time steps')
plt.ylabel('$\sigma$')
plt.grid('true')
plt.title('Fluctuations respect to the initial position')

plt.show()

plt.plot(x, error)
plt.xlabel('# of time steps')
plt.ylabel('error')
plt.grid('true')
plt.title('Statistical uncertainty of the fluctuations')

plt.show()
```

In both random walks (Z3 and R3 lattice) we can note a diffusive behaviour due to the fact that the fluctuations are increasing.
