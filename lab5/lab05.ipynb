```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("r100_ave.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,1000000],[1.5,1.5],label = 'y=1.5')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()

plt.show()
```
