```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("call_final.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,10000],[14.9758,14.9758],label = 'y=14.9758')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()

plt.show()
```
