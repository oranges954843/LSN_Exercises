## **Exercise 1** 

- Let's compute the following 1D integral via Monte Carlo 

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. Results by sampling the integral with a uniform distribution in the interval [0,1] with data blocking for the statistical uncertainties.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("Uniform.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,10000],[1.0,1.0],label = 'y=1')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()

plt.show()
```

2. Results by sampling the integral with a non-uniform distribution in the interval [0,1] with data blocking for the statistical uncertainties. In order to change the probability density to sample [0,1] it has been first made a change in the integrand with an integration per parts.

$$\int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = \int_0^1 \frac{\pi^2}{4}\sin(\pi x/2) x dx$$

After that it is easily found and then normalized the probability density by considering the Taylor expansion of the $\sin$ function in the integrand arrested at the first order

$$p(x) = 3 x ^ 2$$

thus, the integral became

$$I = \int_0^1 \frac{\pi ^2}{12 x}\sin(\pi x/2) dx = 1$$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("NonUniform.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot([0,10000],[1.0,1.0],label = 'y=1')
plt.xlabel('# of throws')
plt.ylabel('<r>')
plt.grid('true')
plt.legend()

plt.show()
```

## **Exercise 2** 
- Let's make 3D Random Walks on a cubic lattice and in the continuum.

1. Random Walk on a cubic lattice with lattice constant $a=1$. It is shown a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$, with its statistical uncertainty, as function of the step $i\in [0,10^2]$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("Z3_lattice.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot(x,x**(1/2), label = 'y=N**(1/2)')
plt.xlabel('# of steps')
plt.ylabel('<RW>')
plt.grid('true')
plt.legend()

plt.show()
```

2. Random Walk in the continuum where at each discrete time the walker makes a step of length equal to $a(=1)$ along a random direction. It is shown a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$, with its statistical uncertainty, as function of the step $i\in [0,10^2]$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("R3_lattice.out", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.plot(x,x**(1/2), label = 'y=N**(1/2)')
plt.xlabel('# of steps')
plt.ylabel('<RW>')
plt.grid('true')
plt.legend()

plt.show()
```
